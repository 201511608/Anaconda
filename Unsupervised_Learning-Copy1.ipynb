```python
#1    #KMeans Cluster  # from sklearn
#     Kmeans remembers the mean of each clusters (the Centroids)
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
####
data=load_iris()
samples=data['data']
species=data['target']

### Train Fit Model
model =KMeans(n_clusters=3)  # Have 3 Clusters  ## [3 species of iris]  ## Put any number and check
model.fit(samples)           # Remembers the fit
labels=model.predict(samples)
print("labels : ",labels)    # setosa  versicolor verginica   {0 1 2}

#1.1  #New_Cluster
new_samples=np.array([[5.7,4.4,1.5,0.4],[6.5,3.0,5.5,1.8],[5.8,2.7,5.1,1.9]])
new_labels=model.predict(new_samples)
print("New_labels : ", new_labels)

#1.2   #Scatter Plots
import matplotlib.pyplot as plt
xs=samples[:,0] # seapel length
ys=samples[:,2] # Peatel length    # change id 0 1 2 3 and check
plt.scatter(xs,ys,c=labels) # labels for colouring
plt.show()
```
