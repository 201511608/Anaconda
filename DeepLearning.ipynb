```python
#7.1
# Gradient Descent
# Learning Rate * Slope/Gradient
# Slope = 2*(ouput-actual_targetoutput)*Inputnodevalue*slopeofActivationfuncion    
                                                #  x2 error  -> slope  2 * X   // if x slope -> 1
# Weights_Update = Weights - Learning_rate * Gradient

# 3 node input 1 output
input_data = np.array([1,2,3])
weights =  np.array([0, 2, 1])
target=0
learning_rate=0.01

preds = (weights * input_data).sum()
error = preds - target
slope = 2  * error * input_data


#7.2 UPDATES
weights_updated = weights - learning_rate * slope
preds_updated = (weights_updated * input_data).sum()
error_updated = preds_updated - target

#Print
print("Slopes :",slope)
print("Error :",error)
print("Error_updaeted :",error_updated)
```

```python
#7.3
# Forword Propogation
# Weight Update Over iterations -> Error to 0
def get_slope(input_data, target, weights):
    # 3 Inputs 3 Weights 1 Output
    preds = (weights * input_data).sum()
    error = preds - target
    slope = 2  * error * input_data
    return slope
def get_mse(input_data, target, weights):
    import math
    preds = (weights * input_data).sum()
    error = preds - target
    e=(math.pow(error,2))
    return e
def weights_Udates(weights,slope,learning_rate):
    weights=weights - learning_rate*slope
    return weights
    
    
#Learning rate and slope  = New Weights
input_data = np.array([0,0,2])
weights =  np.array([1, 1, 6])
learning_rate=0.15
target=-10000



for num in range(20):
    Mean_Square_Error = get_mse(input_data, target, weights)
    slope = get_slope(input_data, target, weights)
    weights =weights_Udates(weights,slope,learning_rate)
    print("MS_Error :",Mean_Square_Error)
```

```python
#8
#Weights Update Via Plot
import numpy as np
import matplotlib.pyplot as plt

input_data = np.array([1, 2, 3])
weights = np.array([0, 2, 1])
target=10000
n_updates = 20
mse_hist = []

# Iterate over the number of updates
for i in range(n_updates):
    # Calculate the slope: slope
    slope = get_slope(input_data, target, weights)
    
    # Update the weights: weights
    weights = weights - 0.01 * slope
    
    # Calculate mse with new weights: mse
    mse = get_mse(input_data, target, weights)
    
    # Append the mse to mse_hist
    mse_hist.append(mse)

# Plot the mse history
plt.plot(mse_hist)
plt.xlabel('Iterations')
plt.ylabel('Mean Squared Error')
plt.show()


    
```

```python
class NuralNetwork_Basic():

    # Functions DeepLearing Basic Forword backwords Activation Gradient\slope Calculations and Plots
    def Mul_Sum(input_data,weights):
        node_value = (input_data * weights).sum()
        return node_value

    def ForwordNet(InputNpArray,WeithsDict):
        #2 2 1
        #two inputs and one hidden layer[with two nodes]   = one output
        import numpy as np
        #Inputs
        input_data=InputNpArray
        #Weights
        weights=WeithsDict

        #HiddenNodes
        node_0_value =Mul_Sum(input_data,weights['node_0']) 
        node_1_value =Mul_Sum(input_data,weights['node_1'])
        #Clubbing Hidden layer
        hidden_layer_values =np.array([node_0_value,node_1_value])
        #OutPutNodes
        output=Mul_Sum(hidden_layer_values,weights['output'])

        #Print
        print("Input data :",InputNpArray)
        print("Hidden Layer :",hidden_layer_values)
        print("OutPut : ", output)
        return output
    def ForwordNet_TanhA(InputNpArray,WeithsDict):
        #2 2 1
        #two inputs and one hidden layer[with two nodes]   = one output
        import numpy as np
        #Inputs
        input_data=InputNpArray
        #Weights
        weights=WeithsDict

        #HiddenNodes
        node_0_value =tanh_A(Mul_Sum(input_data,weights['node_0']))
        node_1_value =tanh_A(Mul_Sum(input_data,weights['node_1']))
        #Clubbing Hidden layer
        hidden_layer_values =np.array([node_0_value,node_1_value])
        #OutPutNodes
        output=Mul_Sum(hidden_layer_values,weights['output'])

        #Print
        print("Input data :",InputNpArray)
        print("Hidden Layer :",hidden_layer_values)
        print("OutPut : ", output)
    def ForwordNet_ReluA(InputNpArray,WeithsDict):
        #2 2 1
        #two inputs and one hidden layer[with two nodes]   = one output
        import numpy as np
        #Inputs
        input_data=InputNpArray
        #Weights
        weights=WeithsDict

        #HiddenNodes
        node_0_value =relu(Mul_Sum(input_data,weights['node_0']))
        node_1_value =relu(Mul_Sum(input_data,weights['node_1']))
        #Clubbing Hidden layer
        hidden_layer_values =np.array([node_0_value,node_1_value])
        #OutPutNodes
        output=Mul_Sum(hidden_layer_values,weights['output'])

        #Print
        print("Input data :",InputNpArray)
        print("Hidden Layer :",hidden_layer_values)
        print("OutPut : ", output)
    def tanh_A(input):
        import numpy as np
        return np.tanh(input)
    def relu(input):
        output =max(0,input)
        return output
    def relu_Numpyarray(numpyarray):
        j=0
        for i in range(len(numpyarray)-1):
            numpyarray[j]=relu(numpyarray[j])
            j=j+1
        return numpyarray



    def ForwordNet_HiddenLayer(InputNpArray,WeithsDict):
        #two inputs and one hidden layer[with two nodes]   = one output
        import numpy as np
        #Inputs
        input_data=InputNpArray
        #Weights
        weights=WeithsDict
        #HiddenNodes
        node_0_value =Mul_Sum(input_data,weights[1]) 
        node_1_value =Mul_Sum(input_data,weights[2])
        output=np.array([node_0_value,node_1_value])
        return output

    def SquareError(numpyarray):
        return numpyarray*numpyarray

    def MeanSquareError(numpyarray):
        numpyarray=SquareError(numpyarray)
        numpyarray=numpyarray.sum()/(len(numpyarray))
        return numpyarray



    def get_slope(input_data, target, weights):
        # 3 1
        # 3 Inputs 3 Weights 1 Output
        preds = (weights * input_data).sum()
        error = preds - target
        slope = 2  * error * input_data
        return slope
    def get_mse(input_data, target, weights):
        # Mean Square Error
        import math
        preds = (weights * input_data).sum()
        error = preds - target
        e=(math.pow(error,2))
        return e
    def weights_Udates(weights,slope,learning_rate):
        weights=weights - learning_rate*slope         # Positive or negative base up on Slope
        return weights


    def Mse_Plot_N31(input_data,weights,target,updates=20):
        # 3 1
        # 3NodeInput 1Output   || 3weights Onnlny

        #
        import numpy as np
        import matplotlib.pyplot as plt

        input_data = input_data # np.array([1, 2, 3])
        weights = weights       #np.array([0, 2, 1])
        target=target
        n_updates = updates
        mse_hist = []

        # Iterate over the number of updates
        for i in range(n_updates):
            # Calculate the slope: slope
            slope = get_slope(input_data, target, weights)

            # Update the weights: weights
            weights = weights - 0.01 * slope

            # Calculate mse with new weights: mse
            mse = get_mse(input_data, target, weights)

            # Append the mse to mse_hist
            mse_hist.append(mse)

        # Plot the mse history
        plt.plot(mse_hist)
        plt.xlabel('Iterations')
        plt.ylabel('Mean Squared Error')
        plt.show()

    def Mse_Plot_N3_31(input_data,weights,Hidden_L_weight_dict,target=0,updates=20):
        #3  3 1
        #3 run once   3   1 iterates
        input_data = input_data # Dosen't work for this data  10->1 check
        weights = weights
        target=target
        updates = updates

        #3 3 1
        weights21 = Hidden_L_weight_dict['weights21']
        weights22 = Hidden_L_weight_dict['weights22']
        weights23 = Hidden_L_weight_dict['weights23']
        h1 = tanh_A((weights21 * input_data).sum())
        h2 = tanh_A((weights22 * input_data).sum())
        h3 = tanh_A((weights23 * input_data).sum())

        input_data=np.array([h1, h2, h3])
        print(input_data)
        Mse_Plot_N31(input_data,weights,target,updates)


    def text_plot(x,y,text,fontsize=10):
        import matplotlib.pyplot as plt
        plt.text(x, y, text, fontsize=fontsize)


    def FNural_2221_Show(input_data,HLW,HL1,HL2,output):
        import math
        import matplotlib.pyplot as plt

        #Circle
        circle1=plt.Circle((0,0+2),.5,color='b',alpha=0.5)
        circle2=plt.Circle((0,4+2),.5,color='b',alpha=0.5)
        circle3=plt.Circle((3,0+2),.5,color='r',alpha=0.5)
        circle4=plt.Circle((3,4+2),.5,color='r',alpha=0.5)
        circle5=plt.Circle((6,0+2),.5,color='r',alpha=0.5)
        circle6=plt.Circle((6,4+2),.5,color='r',alpha=0.5)
        circle7=plt.Circle((9,4),.5,color='b',alpha=0.5)

        plt.xlim(-1,10)
        plt.ylim(-1,10)
        plt.gcf().gca().add_artist(circle1)
        plt.gcf().gca().add_artist(circle2)
        plt.gcf().gca().add_artist(circle3)
        plt.gcf().gca().add_artist(circle4)
        plt.gcf().gca().add_artist(circle5)
        plt.gcf().gca().add_artist(circle6)
        plt.gcf().gca().add_artist(circle7)

        #Text
        text_plot(0-0.25,4+2-0.25,str(input_data[0]),fontsize=10)
        text_plot(0-0.25,0+2-0.25,str(input_data[1]),fontsize=10)

        text_plot(3-0.25,4+2-0.25,str(HL1[0]),fontsize=10)
        text_plot(3-0.25,0+2-0.25,str(HL1[1]),fontsize=10)

        text_plot(6-0.25,4+2-0.25,str(HL2[0]),fontsize=10)
        text_plot(6-0.25,0+2-0.25,str(HL2[1]),fontsize=10)

        text_plot(9-0.25,4-0.25,str(output),fontsize=10)

        #WeightsLines
        #uplines
        plt.arrow(0+0.5,6,2-0.5,0, head_width=0.25, head_length=0.5, fc='k', ec='k')
        plt.arrow(0+0.5,6,2-0.2,-3-0.72, head_width=0.25, head_length=0.5, fc='k', ec='k')

        plt.arrow(3-0.25+0.7,4+2,2-0.5,0, head_width=0.25, head_length=0.5, fc='k', ec='k')
        plt.arrow(3-0.25+0.7,4+2,2-0.2,-3-0.72, head_width=0.25, head_length=0.5, fc='k', ec='k')

        plt.arrow(6-0.20+0.7,4+2,3-1.3,-1.7, head_width=0.25, head_length=0.5, fc='k', ec='k')


        #Downlines
        plt.arrow(0+0.5,0+2,2-0.5,0, head_width=0.25, head_length=0.5, fc='k', ec='k')
        plt.arrow(0+0.5,0+2,2-0.2,-(-3-0.72), head_width=0.25, head_length=0.5, fc='k', ec='k')

        plt.arrow(3+0.5,0+2,2-0.5,0, head_width=0.25, head_length=0.5, fc='k', ec='k')
        plt.arrow(3+0.5,0+2,2-0.2,-(-3-0.72), head_width=0.25, head_length=0.5, fc='k', ec='k')

        plt.arrow(6+0.5,0+2 ,(3-1.3),1.7, head_width=0.25, head_length=0.5, fc='k', ec='k')



        #WeightsText
        text_plot(1.25,4+2+0.25,str(HLW['Node11'][0]),fontsize=12)
        text_plot(1,5,str(HLW['Node12'][0]),fontsize=12)

        text_plot(1.25,2+0.25,str(HLW['Node12'][1]),fontsize=12)
        text_plot(0.5,3,str(HLW['Node11'][1]),fontsize=12)



        text_plot(1.25+3,4+2+0.25,str(HLW['Node21'][0]),fontsize=12)
        text_plot(1+3,5,str(HLW['Node22'][0]),fontsize=12)

        text_plot(1.25+3,2+0.25,str(HLW['Node22'][1]),fontsize=12)
        text_plot(0.5+3,3,str(HLW['Node21'][1]),fontsize=12)



        text_plot(1.25+6,4+2-0.5,str(HLW['Output'][0]),fontsize=12)
        text_plot(1.25+6,2+0.25,str(HLW['Output'][1]),fontsize=12)
        plt.axis('off')
        plt.show()

    def BNural_2221_Show(input_data,HLW,HL1,HL2,output):
        import math
        import matplotlib.pyplot as plt
        #BackWords Nural Nets
        #Circle
        circle1=plt.Circle((0,0+2),.5,color='b',alpha=0.5)
        circle2=plt.Circle((0,4+2),.5,color='b',alpha=0.5)
        circle3=plt.Circle((3,0+2),.5,color='r',alpha=0.5)
        circle4=plt.Circle((3,4+2),.5,color='r',alpha=0.5)
        circle5=plt.Circle((6,0+2),.5,color='r',alpha=0.5)
        circle6=plt.Circle((6,4+2),.5,color='r',alpha=0.5)
        circle7=plt.Circle((9,4),.5,color='b',alpha=0.5)

        plt.xlim(-1,10)
        plt.ylim(-1,10)
        plt.gcf().gca().add_artist(circle1)
        plt.gcf().gca().add_artist(circle2)
        plt.gcf().gca().add_artist(circle3)
        plt.gcf().gca().add_artist(circle4)
        plt.gcf().gca().add_artist(circle5)
        plt.gcf().gca().add_artist(circle6)
        plt.gcf().gca().add_artist(circle7)

        #Text
        text_plot(0-0.25,4+2-0.25,str(input_data[0]),fontsize=10)
        text_plot(0-0.25,0+2-0.25,str(input_data[1]),fontsize=10)

        text_plot(3-0.25,4+2-0.25,str(HL1[0]),fontsize=10)
        text_plot(3-0.25,0+2-0.25,str(HL1[1]),fontsize=10)

        text_plot(6-0.25,4+2-0.25,str(HL2[0]),fontsize=10)
        text_plot(6-0.25,0+2-0.25,str(HL2[1]),fontsize=10)

        text_plot(9-0.25,4-0.25,str(output),fontsize=10)

        #WeightsLines
        #uplines
        plt.arrow(0+0.5+2,6,-(2-0.5),0, head_width=0.25, head_length=0.5, fc='k', ec='k')
        plt.arrow(0+0.5+2-0.2+0.3,6-3-0.72-0.3,-(2-0.2),-(-3-0.72), head_width=0.25, head_length=0.5, fc='k', ec='k')

        plt.arrow(3-0.25+0.7+2,4+2,-(2-0.5),0, head_width=0.25, head_length=0.5, fc='k', ec='k')
        plt.arrow(3-0.25+0.7+2-0.2+0.3,4+2-3-0.72-0.3,-(2-0.2),-(-3-0.72), head_width=0.25, head_length=0.5, fc='k', ec='k')

        plt.arrow(6-0.20+0.7+3-1.3+0.3,4+2-1.7-0.3,-(3-1.3),+1.7, head_width=0.25, head_length=0.5, fc='k', ec='k')


        #Downlines
        plt.arrow(0+0.5+2,0+2,-(2-0.5),0, head_width=0.25, head_length=0.5, fc='k', ec='k')
        plt.arrow(0+0.5+2-0.2+0.3,0+2-(-3-0.72)+0.3,-(2-0.2),+(-3-0.72), head_width=0.25, head_length=0.5, fc='k', ec='k')

        plt.arrow(3+0.5+2,0+2,-(2-0.5),0, head_width=0.25, head_length=0.5, fc='k', ec='k')
        plt.arrow(3+0.5+2-0.2+0.3,0+2-(-3-0.72)+0.3,-(2-0.2),(-3-0.72), head_width=0.25, head_length=0.5, fc='k', ec='k')

        plt.arrow(6+0.5+(3-1.3)+0.3,0+2+1.7+0.3 ,-(3-1.3),-1.7, head_width=0.25, head_length=0.5, fc='k', ec='k')




        #WeightsText
        text_plot(1.25,4+2+0.25,str(HLW['Node11'][0]),fontsize=12)
        text_plot(1,5,str(HLW['Node12'][0]),fontsize=12)

        text_plot(1.25,2+0.25,str(HLW['Node12'][1]),fontsize=12)
        text_plot(0.5,3,str(HLW['Node11'][1]),fontsize=12)



        text_plot(1.25+3,4+2+0.25,str(HLW['Node21'][0]),fontsize=12)
        text_plot(1+3,5,str(HLW['Node22'][0]),fontsize=12)

        text_plot(1.25+3,2+0.25,str(HLW['Node22'][1]),fontsize=12)
        text_plot(0.5+3,3,str(HLW['Node21'][1]),fontsize=12)



        text_plot(1.25+6,4+2-0.5,str(HLW['Output'][0]),fontsize=12)
        text_plot(1.25+6,2+0.25,str(HLW['Output'][1]),fontsize=12)
        plt.axis('off')
        plt.show()    

    def DoFor_Nural_2221_Show(input_data= np.array([0.5, 1]),HLW={'Node11':np.array([2, -2]),
                                                                  'Node12':np.array([4, -1]),
                                                                  'Node21':np.array([0, 1]),
                                                                  'Node22':np.array([2, 3]),
                                                                  'Output':np.array([1,2])    },target=4):
        # Do Forword Nural Network Only
        import matplotlib.pyplot as plt
        #Forword
        Node11=(input_data*HLW['Node11']).sum()
        Node12=(input_data*HLW['Node12']).sum()
        HL1=np.array([Node11,Node12])
        HL1=relu_Numpyarray(HL1)
        Node21=(HL1*HLW['Node21']).sum()
        Node22=(HL1*HLW['Node22']).sum()
        HL2=np.array([Node21,Node22])
        HL2=relu_Numpyarray(HL2)
        output=(HL2*HLW['Output']).sum()
        print("Hidden_L1: ",HL1,"Hidden_L2: ",HL2,"Output: ",output)

        #Plot
        FNural_2221_Show(input_data,HLW,HL1,HL2,output)
        return {"HL1":HL1,"HL2":HL2,"output":output }

    #Forword Backword Show calculaion
    def DoBack_Nural_2221(input_data,HLW,HL1,HL2,Error):
        #Do Back words Nural Networks

        # With Relu activation for Hidden layer  i.e if x>=0 = x  ||  x<0 = 0
        def relu_OP(Node):
            #Relu Slope out put
            if Node >=0:
                return Node
            else :
                return 0

        import copy
        import numpy as np
    #     input_data = np.array([0.5, 1])  
    #     HLW={'Node11':np.array([2, -2]),
    #          'Node12':np.array([4, -1]),
    #          'Node21':np.array([0, 1]),
    #          'Node22':np.array([2, 3]),
    #          'Output':np.array([1,2])    }
    #     HLW_A=copy.deepcopy(HLW)
    #     target=4

        #Forword Propogation # Relu Activation
    #     Node11=(input_data*HLW['Node11']).sum()
    #     Node12=(input_data*HLW['Node12']).sum()
    #     HL1=np.array([Node11,Node12])
    #     HL1=relu_Numpyarray(HL1)
    #     Node21=(HL1*HLW['Node21']).sum()
    #     Node22=(HL1*HLW['Node22']).sum()
    #     HL2=np.array([Node21,Node22])
    #     HL2=relu_Numpyarray(HL2)
    #     output=(HL2*HLW['Output']).sum()
    #     print("Hidden L1: ",HL1,"\nHidden L2: ",HL2,"\nOutput :   ",output)
    #     DoFor_Nural_2221_Show(input_data,HLW,target)
    #     #Error
    #     Error=output-target
    #     print("Error : ",Error)


        #Backword Propogation # Relu Activation
        #3
        #print("\nBackPropogation")
    #     HLW['Output']=np.array([1,12])

        #Update weights
        HLW['Output'][0]=HL2[0]*(2*Error)*(1)
        HLW['Output'][1]=HL2[1]*(2*Error)*(1)
        Node21=HLW['Output'][0]
        Node22=HLW['Output'][1]
        HL2_N=np.array([Node21,Node22])
        #print("Gradient HL2_N: ",HL2_N)
        #BNural_2221_Show(input_data,HLW,HL1,HL2_N,output)
        #2
        HLW['Node21'][0]=relu_OP(Node21*HL1[0])#1
        HLW['Node22'][0]=relu_OP(Node22*HL1[0])#2

        HLW['Node21'][1]=relu_OP(Node21*HL1[1])#1
        HLW['Node22'][1]=relu_OP(Node22*HL1[1])#2
        #print(HLW['Node21'][0],HLW['Node22'][0],HLW['Node21'][1],HLW['Node22'][1])
        Node11=HLW['Node21'][0] + HLW['Node22'][0]
        Node12=HLW['Node21'][1] + HLW['Node22'][1]
        HL1_N=np.array([Node11,Node12])
        #print("Gradient HL1_N: ",HL1_N)
        #BNural_2221_Show(input_data,HLW,HL1_N,HL2_N,output)

        #1
        HLW['Node11'][0]=relu_OP(Node11*input_data[0])#1
        HLW['Node12'][0]=relu_OP(Node12*input_data[0])#2

        HLW['Node11'][1]=relu_OP(Node11*input_data[1])#1
        HLW['Node12'][1]=relu_OP(Node12*input_data[1])#2
        #print(HLW['Node11'][0],HLW['Node12'][0],HLW['Node11'][1],HLW['Node12'][1])
        Node01=HLW['Node11'][0] + HLW['Node12'][0]
        Node02=HLW['Node11'][1] + HLW['Node12'][1]
        HL0_N=np.array([Node01,Node02])
        #print("Gradient HL0_N: ",HL0_N)
        #BNural_2221_Show(HL0_N,HLW,HL1_N,HL2_N,output)
        return {"HLW":HLW,"HL0_N":HL0_N,"HL1_N":HL1_N,"HL2_N":HL2_N}

    # FBDo_Nural_2221  ::  Do Forword and Backworkds Propagation and return Gradient 
    def FBDo_Nural_2221(self,input_data= np.array([0.5, 1]),HLW={'Node11':np.array([2, -2]),
                                                              'Node12':np.array([4, -1]),
                                                              'Node21':np.array([0, 1]),
                                                              'Node22':np.array([2, 3]),
                                                              'Output':np.array([1,2])    },target=4):
        #Do Forword and Backworkds Propagation and return Gradient

        F=DoFor_Nural_2221_Show(input_data,HLW,target)
        error=F["output"]-target
        print("Error= Obtain - target =",F["output"]," - ",target," = ",error)
        B=DoBack_Nural_2221(input_data,HLW,F["HL1"],F["HL2"],error)
        BNural_2221_Show(B["HL0_N"],B["HLW"],B["HL1_N"],B["HL2_N"],error) #error or F["output"]
        return B["HLW"]

    def FB_Nural_2221_iteration(self,input_data= np.array([0.5, 1]),
        HLW={'Node11':np.array([2, -2]),
          'Node12':np.array([4, -1]),
          'Node21':np.array([0, 1]),
          'Node22':np.array([2, 3]),
          'Output':np.array([1,2])    },
         target=4,
        Learning_rate=0.01,iteration=10)  :
        import copy
        import numpy as np

        HLW_u=copy.deepcopy(HLW)
        FB_Gradient=FBDo_Nural_2221(input_data,copy.deepcopy(HLW),target)
        x=0
        while (x<iteration):
            for key in HLW_u:
                HLW_u[key]=HLW_u[key]-Learning_rate*FB_Gradient[key]
            FB_Gradient=FBDo_Nural_2221(input_data,copy.deepcopy(HLW_u),target)
            x=x+1

```
