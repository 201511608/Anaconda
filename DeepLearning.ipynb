```python
#7
# Gradient Descent
# Learning Rate * Slope/Gradient
# Slope = 2*(ouput-actual_targetoutput)*Inputnodevalue*slopeofActivationfuncion    
                                                #  x2 error  -> slope  2 * X   // if x slope -> 1
# Weights_Update = Weights - Learning_rate * Gradient

# 3 node input 1 output
input_data = np.array([1,2,3])
weights =  np.array([0, 2, 1])
target=0
learning_rate=0.01

preds = (weights * input_data).sum()
error = preds - target
slope = 2  * error * input_data


#7.2 UPDATES
weights_updated = weights - learning_rate * slope
preds_updated = (weights_updated * input_data).sum()
error_updated = preds_updated - target

#Print
print("Slopes :",slope)
print("Error :",error)
print("Error_updaeted :",error_updated)
```

```python
#7.3
# Forword Propogation
# Weight Update Over iterations -> Error to 0
def get_slope(input_data, target, weights):
    # 3 Inputs 3 Weights 1 Output
    preds = (weights * input_data).sum()
    error = preds - target
    slope = 2  * error * input_data
    return slope
def get_mse(input_data, target, weights):
    import math
    preds = (weights * input_data).sum()
    error = preds - target
    e=(math.pow(error,2))
    return e
def weights_Udates(weights,slope,learning_rate):
    weights=weights - learning_rate*slope
    return weights
    
    
#Learning rate and slope  = New Weights
input_data = np.array([0,0,2])
weights =  np.array([1, 1, 6])
learning_rate=0.15
target=-10000



for num in range(20):
    Mean_Square_Error = get_mse(input_data, target, weights)
    slope = get_slope(input_data, target, weights)
    weights =weights_Udates(weights,slope,learning_rate)
    print("MS_Error :",Mean_Square_Error)
```

```python
#8
#Weights Update Via Plot
import numpy as np
import matplotlib.pyplot as plt

input_data = np.array([1, 2, 3])
weights = np.array([0, 2, 1])
target=10000
n_updates = 20
mse_hist = []

# Iterate over the number of updates
for i in range(n_updates):
    # Calculate the slope: slope
    slope = get_slope(input_data, target, weights)
    
    # Update the weights: weights
    weights = weights - 0.01 * slope
    
    # Calculate mse with new weights: mse
    mse = get_mse(input_data, target, weights)
    
    # Append the mse to mse_hist
    mse_hist.append(mse)

# Plot the mse history
plt.plot(mse_hist)
plt.xlabel('Iterations')
plt.ylabel('Mean Squared Error')
plt.show()


    
```

```python
def Mul_Sum(input_data,weights):
    node_value = (input_data * weights).sum()
    return node_value

def ForwordNet(InputNpArray,WeithsDict):
    #2 2 1
    #two inputs and one hidden layer[with two nodes]   = one output
    import numpy as np
    #Inputs
    input_data=InputNpArray
    #Weights
    weights=WeithsDict
    
    #HiddenNodes
    node_0_value =Mul_Sum(input_data,weights['node_0']) 
    node_1_value =Mul_Sum(input_data,weights['node_1'])
    #Clubbing Hidden layer
    hidden_layer_values =np.array([node_0_value,node_1_value])
    #OutPutNodes
    output=Mul_Sum(hidden_layer_values,weights['output'])

    #Print
    print("Input data :",InputNpArray)
    print("Hidden Layer :",hidden_layer_values)
    print("OutPut : ", output)
    return output
def ForwordNet_TanhA(InputNpArray,WeithsDict):
    #2 2 1
    #two inputs and one hidden layer[with two nodes]   = one output
    import numpy as np
    #Inputs
    input_data=InputNpArray
    #Weights
    weights=WeithsDict
    
    #HiddenNodes
    node_0_value =tanh_A(Mul_Sum(input_data,weights['node_0']))
    node_1_value =tanh_A(Mul_Sum(input_data,weights['node_1']))
    #Clubbing Hidden layer
    hidden_layer_values =np.array([node_0_value,node_1_value])
    #OutPutNodes
    output=Mul_Sum(hidden_layer_values,weights['output'])

    #Print
    print("Input data :",InputNpArray)
    print("Hidden Layer :",hidden_layer_values)
    print("OutPut : ", output)
def ForwordNet_ReluA(InputNpArray,WeithsDict):
    #2 2 1
    #two inputs and one hidden layer[with two nodes]   = one output
    import numpy as np
    #Inputs
    input_data=InputNpArray
    #Weights
    weights=WeithsDict
    
    #HiddenNodes
    node_0_value =relu(Mul_Sum(input_data,weights['node_0']))
    node_1_value =relu(Mul_Sum(input_data,weights['node_1']))
    #Clubbing Hidden layer
    hidden_layer_values =np.array([node_0_value,node_1_value])
    #OutPutNodes
    output=Mul_Sum(hidden_layer_values,weights['output'])

    #Print
    print("Input data :",InputNpArray)
    print("Hidden Layer :",hidden_layer_values)
    print("OutPut : ", output)
def tanh_A(input):
    import numpy as np
    return np.tanh(input)
def relu(input):
    output =max(0,input)
    return output
def relu_Numpyarray(numpyarray):
    j=0
    for i in range(len(numpyarray)-1):
        numpyarray[j]=relu(numpyarray[j])
        j=j+1
    return numpyarray



def ForwordNet_HiddenLayer(InputNpArray,WeithsDict):
    #two inputs and one hidden layer[with two nodes]   = one output
    import numpy as np
    #Inputs
    input_data=InputNpArray
    #Weights
    weights=WeithsDict
    #HiddenNodes
    node_0_value =Mul_Sum(input_data,weights[1]) 
    node_1_value =Mul_Sum(input_data,weights[2])
    output=np.array([node_0_value,node_1_value])
    return output

def SquareError(numpyarray):
    return numpyarray*numpyarray

def MeanSquareError(numpyarray):
    numpyarray=SquareError(numpyarray)
    numpyarray=numpyarray.sum()/(len(numpyarray))
    return numpyarray



def get_slope(input_data, target, weights):
    # 3 1
    # 3 Inputs 3 Weights 1 Output
    preds = (weights * input_data).sum()
    error = preds - target
    slope = 2  * error * input_data
    return slope
def get_mse(input_data, target, weights):
    # Mean Square Error
    import math
    preds = (weights * input_data).sum()
    error = preds - target
    e=(math.pow(error,2))
    return e
def weights_Udates(weights,slope,learning_rate):
    weights=weights - learning_rate*slope         # Positive or negative base up on Slope
    return weights
    
    
def Mse_Plot_N31(input_data,weights,target,updates=20):
    # 3 1
    # 3NodeInput 1Output   || 3weights Onnlny
    
    #
    import numpy as np
    import matplotlib.pyplot as plt

    input_data = input_data # np.array([1, 2, 3])
    weights = weights       #np.array([0, 2, 1])
    target=target
    n_updates = updates
    mse_hist = []

    # Iterate over the number of updates
    for i in range(n_updates):
        # Calculate the slope: slope
        slope = get_slope(input_data, target, weights)

        # Update the weights: weights
        weights = weights - 0.01 * slope

        # Calculate mse with new weights: mse
        mse = get_mse(input_data, target, weights)

        # Append the mse to mse_hist
        mse_hist.append(mse)

    # Plot the mse history
    plt.plot(mse_hist)
    plt.xlabel('Iterations')
    plt.ylabel('Mean Squared Error')
    plt.show()

def Mse_Plot_N3_31(input_data,weights,Hidden_L_weight_dict,target=0,updates=20):
    #3  3 1
    #3 run once   3   1 iterates
    input_data = input_data # Dosen't work for this data  10->1 check
    weights = weights
    target=target
    updates = updates

    #3 3 1
    weights21 = Hidden_L_weight_dict['weights21']
    weights22 = Hidden_L_weight_dict['weights22']
    weights23 = Hidden_L_weight_dict['weights23']
    h1 = tanh_A((weights21 * input_data).sum())
    h2 = tanh_A((weights22 * input_data).sum())
    h3 = tanh_A((weights23 * input_data).sum())

    input_data=np.array([h1, h2, h3])
    print(input_data)
    Mse_Plot_N31(input_data,weights,target,updates)
```
