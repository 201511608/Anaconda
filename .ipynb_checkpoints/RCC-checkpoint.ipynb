```python
# 1
# Nutral Axis Data
# Single Reinforcement Beam
fy=[250,415,550]
fck=[15,20,25,30,35,40,45,50,60,70]  
b=230
d=300
file=open('SingeReinforcementNutralAxis.csv','w+')
line_count=0

for i_b in range(b,750+10,10):  
#     print(i_b)
    for i_d in range(d,750+10,10):# depth taken from 300 to 750
        for i_fck in fck:
            for i_fy in fy:
                for i_Ast in range(100, 1500+10,10):
                    xu=(0.87*i_fy*i_Ast)/(0.36*i_fck*i_Ast)
                    file.write(str(i_b)+","+str(i_d)+","+str(i_fck)+","+str(i_fy)+","+str(i_Ast)+","+str(xu))
                    file.write("\n")
                    line_count=line_count+1
    
file.close()
print(line_count)
```

```python
# 2
# Nutral Axis Data
# Double Reinforcement Beam
# Data Making
def DoubleReinforcementNutralAxis(fck=30,fst=415,d_=60,d=550,b=350,Ast=4310,Asc=1690):
    import numpy as np
    if fst == 415:
        strain=[0.00144,0.00163,0.00192,0.00241,0.00276,0.00380]
        stress=[288,    306,    324   ,342    ,351     ,360]
    elif fst ==500:
        strain=[0.00174,0.00195,0.00226,0.00277,0.00312,0.00417]
        stress=[347,    369,    391   ,413    ,423     ,434]


    x=0.1  # Assume the Start of nutral axixnutral axis   # Assume 0 to d


    import math
    F_diff=[]
    minimum=999999999999999999999999999999
    nutral_axis=0
    Tension=0
    Compression=0
    while x<= d:
        esc=0.0035*(1- d_/x)
        fsc=np.interp(esc, strain, stress)

        Cs=fsc*Asc
        Cc=0.36*fck*b*x
        C=Cc+Cs

        # Tensile Steel Yields
        est=0.0035*(d-x)/x
        if est >=0.0035: 
            T=0.87*fst*Ast
        else:
            T=fst*Ast

        F_diff.append(abs(T-C))


        if abs(T-C)<=minimum:   # 
            minimum=abs(T-C)
            nutral_axis= x
            Tension=T
            Compression=C


            
    #         break       
        x=0.001+x     
    #####
    return nutral_axis,Tension,Compression



def  DoubleReinforcementNutralAxis_Type2(fck=30,fst=415,d_=60,d=550,b=350,Ast=4310,Asc=1690):
    # Modified  Ast-Asc<=1000
    import math
    if abs(Ast-Asc)<=1000:
        #1
        fst=fst
        if fst == 415:
            strain=[0.00144,0.00163,0.00192,0.00241,0.00276,0.00380]
            stress=[288,    306,    324   ,342    ,351     ,360]
        elif fst ==500:
            strain=[0.00174,0.00195,0.00226,0.00277,0.00312,0.00417]
            stress=[347,    369,    391   ,413    ,423     ,434]

        fck=fck
        d_=d_
        d=d
        b=b
        Ast=Ast 
        Asc=Asc
        x=0.00001  # Assume the Start of nutral axixnutral axis   # Assume 0 to d
        F_diff=[]
        minimum=9999999999999999999999999
        nutral_axis=0.00001
        Tension=0
        Compression=0
        while x<= d:
            esc=0.0035*(1- d_/x)
            fsc=np.interp(esc, strain, stress)

            Cs=fsc*Asc
            Cc=0.36*fck*b*x
            C=Cc+Cs

            # Tensile Steel Yields
            est=0.0035*(d-x)/x
            if est >=0.0035: 
                T=0.87*fst*Ast
            else:
                T=fst*Ast



            if abs(T-C)<=minimum:   # 
                minimum=abs(T-C)
                nutral_axis= x
                Tension=T
                Compression=C
                #break       
            x=0.001+x 
            return nutral_axis,Tension,Compression
    ###
#                             print(fck, fst, b,d,d_,Ast,Asc,nutral_axis, Tension,Compression,Tension-Compression)
                              #Store All data later filter this data !_! 


##
# DoubleReinforcementNutralAxis()
# DoubleReinforcementNutralAxis_Type2()
```

```python
#3
import numpy as np
import pandas as pd

# Nutral Nets Train.
df=pd.read_csv('Data_files/DoubleReinforcementNutralAxis1.csv', sep=',',header=None)
df= df[df[7] > 10]  # Data frame Filter   nutral axis condition 

X=df.drop([7,8,9,10],axis=1).values
y=df[7].values
```

```python
#3
import keras
from keras.layers import Dense
from keras.models import Sequential

#Import Data
predictors=X
target=y
n_cols = predictors.shape[1]

# Model
model = Sequential()
model.add(Dense(100, activation= 'relu',input_shape=(n_cols,))) 
model.add(Dense(100, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(25, activation='relu'))
model.add(Dense(1))

#compile
model.compile(optimizer='adam', loss='mean_squared_error')

#Fit
# model.fit(predictors, target)

from keras.callbacks import EarlyStopping

early_stopping_monitor = EarlyStopping(patience=6)
model.fit(predictors, target, epochs=20,callbacks = [early_stopping_monitor])
```
