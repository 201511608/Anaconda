```python

class Cminmax:
    
    
    Game_Matrix=np.chararray((3,3),unicode=True)
    dic = {"row":-1, "col":-1}

    player ='X'
    opponent ='O'

    def evaluate(b=Game_Matrix):
        # X or O victory check and return +10  or -10

        #Row victory
        for row in range(0,3):
            if b[row][0]==b[row][1] and b[row][1]==b[row][2]:
                if b[row][0]=='X':
                    return +10
                elif b[row][0]=='O':
                    return -10


         #Col victory
        for col in range(0,3):
            if b[0][col]==b[1][col] and b[1][col]==b[2][col]:
                if b[0][col]=='X':
                    return +10
                elif b[0][col]=='O':
                    return -10         

         #Diagonal victory      
        if (b[0][0]==b[1][1] and b[1][1]==b[2][2]):  
            if (b[0][0]=='X'):
                return +10 
            elif (b[0][0]=='O'):
                return -10    
        if (b[0][2]==b[1][1] and b[1][1]==b[2][0]):  
            if (b[0][2]=='X'):
                return +10 
            elif (b[0][2]=='O'):
                return -10 
        return 0



    def ismovesleft(b=Game_Matrix):   
        for i in range(0,3):       
             for j in range(0,3):
                    if (b[i][j]==''):
                        return True
        return False              



    
    def minmax(Game_Matrix,depth,ismax):

        board=Game_Matrix
        score=Cminmax.evaluate(board)

        if (score == 10):
            return score
        if (score == -10):
            return score
        if (Cminmax.ismovesleft(board)==False):
            return 0

        # opponent player play Moves best move output     Min Max
        if (ismax):
            best=-1000 # default to get max from minmax
            for i in range(0,3):
                for j in range(0,3):                 
                    if (board[i][j]==''):  
                        board[i][j]=player
                        best=max(best,Cminmax.minmax(board,depth+1,not ismax))
                        #undo move
                        board[i][j]=''
                     #   print("in")
            return best
        else:
            best=1000 #any max value to get min using minmax min
            for i in range(0,3):
                for j in range(0,3):
                    if (board[i][j]==''): 
                        board[i][j]=opponent
                        best=min(best,Cminmax.minmax(board,depth+1,not ismax))

                        #undo move
                        board[i][j]=''
                       # print("out")
            return best

 

    def findbestmove(board=Game_Matrix):
        bestval=-1000
        bestmove= dic

        #default Null ids
        bestmove["row"]=-1
        bestmove["col"]=-1

        for i in range(0,3):
            for j in range(0,3):
                    if (board[i][j]==''):
                        board[i][j]=player
                        moveval=Cminmax.minmax(board,0,False)
                        board[i][j]=''

                        if (moveval>bestval):
                            bestmove["row"]=i
                            bestmove["col"]=j                        
                            bestval=moveval


        print("Best val :",bestval)
        return bestmove      
    
    def convertion_x_o(board=np.chararray((3,3),unicode=True)):
        
        for i in range(0,3):
            for j in range(0,3):
                if (board[i][j]!=''): 
                    if (board[i][j]=='X'):
                        board[i][j]='O'
                    else:
                        board[i][j]='X'
        return board
    
    
    # Overload minmax function on intelligence [On scale 1-9]
   
    def minmax_1(Game_Matrix,depth,ismax,intelligence):
        board=Game_Matrix
        score=Cminmax.evaluate(board)

        if (score == 10):
            return score
        if (score == -10):
            return score
        if (Cminmax.ismovesleft(board)==False):
            return 0
        if (depth==intelligence):
            return 0

        # opponent player play Moves best move output     Min Max
        if (ismax):
            best=-1000 # default to get max from minmax
            for i in range(0,3):
                for j in range(0,3):                 
                    if (board[i][j]==''):  
                        board[i][j]=player
                        best=max(best,Cminmax.minmax_1(board,depth+1,not ismax,intelligence))
                        #undo move
                        board[i][j]=''
                     #   print("in")
            return best
        else:
            best=1000 #any max value to get min using minmax min
            for i in range(0,3):
                for j in range(0,3):
                    if (board[i][j]==''): 
                        board[i][j]=opponent
                        best=min(best,Cminmax.minmax_1(board,depth+1,not ismax,intelligence))

                        #undo move
                        board[i][j]=''
                       # print("out")
            return best
        
    def findbestmove_1(board=Game_Matrix,intelligence=1):
        bestval=-1000
        bestmove= dic

        #default Null ids
        bestmove["row"]=-1
        bestmove["col"]=-1

        for i in range(0,3):
            for j in range(0,3):
                    if (board[i][j]==''):
                        board[i][j]=player
                        moveval=Cminmax.minmax_1(board,0,False,intelligence)
                        board[i][j]=''

                        if (moveval>bestval):
                            bestmove["row"]=i
                            bestmove["col"]=j                        
                            bestval=moveval


        # print("Best val :",bestval)
        return bestmove      


def convertion_data_num( board=np.chararray((3,3),unicode=True)):
    board
    OP=np.arange(9).reshape(3,3)
    OP[:]=999
    for i in range(0,3):
        for j in range(0,3):
            if (board[i][j]=='X'): 
                OP[i][j]=1
            if (board[i][j]=='O'): 
                OP[i][j]=3
            if (board[i][j]==''): 
                OP[i][j]=0
    return OP

def convertion_best_1_9(X,Y):
    if (X==0 and Y==0):
        return 1
    elif(X==0 and Y==1):
        return 2
    elif(X==0 and Y==2):
        return 3
    elif(X==1 and Y==0):
        return 4
    elif(X==1 and Y==1):
        return 5
    elif(X==1 and Y==2):
        return 6
    elif(X==2 and Y==0):
        return 7
    elif(X==2 and Y==1):
        return 8
    elif(X==2 and Y==2):
        return 9

        
    

      
```
