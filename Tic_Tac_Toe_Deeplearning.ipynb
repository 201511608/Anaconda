```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV



# from keras.layers import Dense
# pred=np.loadtxt('MinMax_Data1.csv',delimiter=',')
data=pd.read_csv('MinMax_Data1.csv')
print (data.head())
X=data.drop(['OUTCOME','STATUS_X','STATUS_O','INTILLEGENCE_X','INTILLEGENCE_O'],axis=1).values
Y=data['OUTCOME'].values
# print(data.head())
# print(X)
# print(Y)
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=42)
reg=LogisticRegression()


reg.fit(X_train,Y_train)
y_pred=reg.predict(X_test)
print(y_pred)
print(Y_test)
reg.score(X_test,Y_test)


```

```python
def convertion_data_num( board=np.chararray((3,3),unicode=True)):
    board
    OP=np.arange(9).reshape(3,3)
    OP[:]=999
    for i in range(0,3):
        for j in range(0,3):
            if (board[i][j]=='X'): 
                OP[i][j]=1
            if (board[i][j]=='O'): 
                OP[i][j]=3
            if (board[i][j]==''): 
                OP[i][j]=0
    return OP
def convertion_best_1_9_X_Y(X):
    if (X==1):
        return np.array([0,0])
    elif(X==2):
        return np.array([0,1])
    elif(X==3):
        return np.array([0,2])
    elif (X==4):
        return np.array([1,0])
    elif(X==5):
        return np.array([1,1])
    elif(X==6):
        return np.array([1,2])
    elif (X==7):
        return np.array([2,0])
    elif(X==8):
        return np.array([2,1])
    elif(X==9):
        return np.array([2,2])
```
