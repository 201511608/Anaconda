```python
#4.1
# 'Democrat' or 'Republican'
#  Voting
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
df=pd.read_csv('US_Voting_Data.csv')        # Must do Preprossing Before using the Data
print(df.head())
print(df.shape)
print(df.info())
print(df.describe())
print('Do this in 4th chapter')


# # Preprocessing Steps
df.party.replace('?',np.nan, inplace=True)
df.infants.replace('?',np.nan, inplace=True)
df.water.replace('?',np.nan, inplace=True)
df.budget.replace('?',np.nan, inplace=True)
df.physician.replace('?',np.nan, inplace=True)
df.salvador.replace('?',np.nan, inplace=True)
df.religious.replace('?',np.nan, inplace=True)
df.satellite.replace('?',np.nan, inplace=True)

df.aid.replace('?',np.nan, inplace=True)
df.missile.replace('?',np.nan, inplace=True)
df.immigration.replace('?',np.nan, inplace=True)
df.synfuels.replace('?',np.nan, inplace=True)
df.education.replace('?',np.nan, inplace=True)
df.superfund.replace('?',np.nan, inplace=True)
df.crime.replace('?',np.nan, inplace=True)

df.duty_free_exports.replace('?',np.nan, inplace=True)
df.eaa_rsa.replace('?',np.nan, inplace=True)

df.replace(('y', 'n'), (1, 0), inplace=True)  # Total DataFrame replace  y/n -> 1/0
 
df.dropna(axis=0, how='any', inplace = True)# Remove all nan

print("\ndf.head() :\n",df.head())
print("\ndf.info() :\n",df.info())



#Plot

plt.style.use('ggplot')
features=df.loc[:, df.columns != 'party']
labels=df.party.replace(('democrat', 'republican'), (1, 0))
_=pd.scatter_matrix(features,c=labels,figsize = [25,25],s=150,marker ='D') 
plt.show()

```

```python
# 5
# Classification
# K-Nearest Neighbors
        # k 5 -> max of 5 points
        # Make a Sets of Decision Bountaries
# On iris data
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets                       #Import Datasets  

iris = datasets.load_iris()

knn= KNeighborsClassifier(n_neighbors=6)

knn.fit(iris['data'],iris['target'])   # inputs as Numpy/pandasDataFrame

prediction = knn.predict(iris['data'])

print('Prediction :',prediction)
print('target',iris.target)
```
